```python
def convert_nD(arr):
    arr = list(arr)
    m = len(arr)
    n = int(np.sqrt(m))
    arr_new = [arr[n*i:n*(i+1)] for i in range(0,int(n))]    
    #print(arr_new)
    for x in range(0,n):
        arr_new[x] = [np.array([0,128,0]) if i=='A' else i for i in arr_new[x]]
        arr_new[x] = [np.array([255,0,0]) if i=='T' else i for i in arr_new[x]]
        arr_new[x] = [np.array([255,165,0]) if i=='C' else i for i in arr_new[x]]
        arr_new[x] = [np.array([1,0,255]) if i=='G' else i for i in arr_new[x]]
    return arr_new

def convert_split_A(arr):
    arr = list(arr)
    m = len(arr)
    n = int(np.sqrt(m))
    arr_new = [arr[n*i:n*(i+1)] for i in range(0,int(n))]    
    #print(arr_new)
    for x in range(0,n):
        arr_new[x] = [np.array([0,128,0]) if i=='A' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='T' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='C' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='G' else i for i in arr_new[x]]
    return arr_new

def convert_split_T(arr):
    arr = list(arr)
    m = len(arr)
    n = int(np.sqrt(m))
    arr_new = [arr[n*i:n*(i+1)] for i in range(0,int(n))]    
    #print(arr_new)
    for x in range(0,n):
        arr_new[x] = [np.array([0,0,0]) if i=='A' else i for i in arr_new[x]]
        arr_new[x] = [np.array([255,0,0]) if i=='T' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='C' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='G' else i for i in arr_new[x]]
    return arr_new

def convert_split_C(arr):
    arr = list(arr)
    m = len(arr)
    n = int(np.sqrt(m))
    arr_new = [arr[n*i:n*(i+1)] for i in range(0,int(n))]    
    #print(arr_new)
    for x in range(0,n):
        arr_new[x] = [np.array([0,0,0]) if i=='A' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='T' else i for i in arr_new[x]]
        arr_new[x] = [np.array([255,165,0]) if i=='C' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='G' else i for i in arr_new[x]]
    return arr_new

def convert_split_G(arr):
    arr = list(arr)
    m = len(arr)
    n = int(np.sqrt(m))
    arr_new = [arr[n*i:n*(i+1)] for i in range(0,int(n))]    
    #print(arr_new)
    for x in range(0,n):
        arr_new[x] = [np.array([0,0,0]) if i=='A' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='T' else i for i in arr_new[x]]
        arr_new[x] = [np.array([0,0,0]) if i=='C' else i for i in arr_new[x]]
        arr_new[x] = [np.array([1,0,255]) if i=='G' else i for i in arr_new[x]]
    return arr_new
```

```python
def wholematrix(seq,ref):
    sub1 = np.asarray(convert_split_A(seq)) & np.asarray(convert_split_A(ref))
    sub2 = np.asarray(convert_split_T(seq)) & np.asarray(convert_split_T(ref))
    sub3 = np.asarray(convert_split_C(seq)) & np.asarray(convert_split_C(ref))
    sub4 = np.asarray(convert_split_G(seq)) & np.asarray(convert_split_G(ref))
    sub = sub1 + sub2 +sub3 + sub4
    return sub
```

### Defining the Leveinshtein distance

```python
def Leivenshtein_Distance(str1, str2):
    m = len(str1)
    n = len(str2)

    dp = [[0 for x in range(n + 1)] for x in range(m + 1)]
    
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j    # Min. operations = j
            elif j == 0:
                dp[i][j] = i    # Min. operations = i
 
            elif str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]

            else:
                dp[i][j] = 1 + min(dp[i][j-1],        # Insert
                                   dp[i-1][j],        # Remove
                                   dp[i-1][j-1])    # Replace
                
    return dp[m][n]

dp = Leivenshtein_Distance(ReferenceSequence, ActualSequence)
```

```python
print(f"Number of mutations in the genome: {dp}")
```

### Defining the Hamming distance

```python
def Hamming_distance(human_gene,mutated_gene):
    n = len(human_gene)
    r = len(mutated_gene)
    hamming_distance = 0
    indexes = []
    
    for i in range(0,n):
        if(human_gene[i]!=mutated_gene[i]):
            hamming_distance = hamming_distance + 1
            indexes.append(i)
            print(f"Mutation detected at index position{i}")
            
    return hamming_distance

distance = Hamming_distance(ReferenceSequence, ActualSequence)
print("Number of mutations in the genome: " + str(distance))
print(f"Percentage of mutations: {distance*100/len(ActualSequence)}")
```
